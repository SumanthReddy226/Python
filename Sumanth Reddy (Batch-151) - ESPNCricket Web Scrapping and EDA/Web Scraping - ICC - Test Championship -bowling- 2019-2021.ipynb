```python
# Loading required libraries

import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

import warnings
warnings.filterwarnings('ignore')

import time
```

```python
# For scrapping the webpage

URL9 = 'https://stats.espncricinfo.com/ci/engine/records/bowling/most_wickets_career.html?id=804;type=trophy'
```

```python
# sends a GET request to the url.

page9 = requests.get(URL9)
page9.status_code


# Obtaining the HTML code from the url



page9.text
```

```python
# For formatting the content of the page and for identifying the html tags

page9_content = page9.text
soup9 = BeautifulSoup(page9_content)
soup9
```

```python
# find PageElements that match the given criteria.

x9 = soup9.find('table')
x9.text
```

```python
# Placing the players stats as a list of items

a = []
for j in range(0,48):
    for i in soup9.find_all('tr')[j].find_all("td"):
        a.append(i.text)

a
```

```python
# Arranging the players stats in the columns by performing indexing

Name = a[0::15]
Span = a[1::15]
Matches= a[2::15]
innings = a[3::15]
Overs = a[4::15]
Maidens = a[5::15]
Runs_concede = a[6::15]
Wickets = a[7::15]
Best_Bowling_innings= a[8::15]
Best_Bowling_match =a[9::15]
Average = a[10::15]
Economy= a[11::15]
Strike_rate =a[12::15]
fifer_5=a[13::15]
ten_wickets = a[14::15]
```

```python
# Storing the columns in the form of a dataframe

Bowling_stats= list(zip(Name,Span,Matches,innings,Overs,Maidens,Runs_concede,Wickets,Best_Bowling_innings,Best_Bowling_match,Average,Economy,Strike_rate,fifer_5,ten_wickets))
Bowling_stats= pd.DataFrame(Bowling_stats, columns = ['Name','Span','Matches','innings','Overs','Maidens','Runs_concede','Wickets','Best_Bowling_innings','Best_Bowling_match','Average','Economy','Strike_rate','fifer_5','ten_wickets'])

Icc_WTC_Bowling_Stats = Bowling_stats
```
